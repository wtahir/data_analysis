```python
def filter_field(csvPath):
    df = pd.read_csv(csvPath, sep=',')
    df['gt'].fillna("This is an empty string for gt", inplace = True)
    df['pred'].fillna("This is an empty string for pred", inplace = True)
    df = df.where(df['gt'] != df['pred'])
    df_group = df.groupby(df['field']).count()
    df_group.drop(['page' ,'gt', 'pred', 'levenshtein_distance', 'doc_conf', 'entity_conf', 'corrected'], axis=1, inplace=True)
    df_group.columns = ['count']
    return df_group.to_csv('field_on_count.csv', sep=',')
```

```python
filter_field('entities.csv')
```
