```python
import pagexml
import pandas as pd
import glob
import os
from tqdm import tqdm
```

```python
df = pd.read_csv('VIE001/coverage_check_question_annotation.csv', encoding='utf8')

for index, row in df.iterrows():
    q1 = row['Größe des Glases (m²)']
    q2 = row['Zustand']
    q3 = row['Schadenort']
    q4 = row['Situation, in der der Schaden entstanden ist']
    q5 = row['Beschädigtes Fenster war schon eingebaut?']
    string = os.path.basename(row['link'])
    files = glob.glob('VIE001/rejected_nicht_versichertes_objekt/*.xml')
    for xml in tqdm(files):
        pxml = pagexml.PageXML(xml)
        file_name = pxml.getValue('/_:PcGts/_:Property[@key="uuid"]/@value')
        node = pxml.selectNth('//_:PcGts', 0)
        if string == file_name:
            pxml.setProperty(node, 'question1', q1)
            pxml.setProperty(node, 'question2', q2)
            pxml.setProperty(node, 'question3', q3)
            pxml.setProperty(node, 'question4', q4)
            pxml.setProperty(node, 'question5', q5)
            pxml.write(xml)
        
```
