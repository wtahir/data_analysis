```python
import pandas as pd
import csv
import glob
import json
import numpy as np
import os
from shapely.geometry import Polygon
from shapely.geometry import point
```

# Heading 1
This notebook gives 3 csv files based upon predicted results with an array of results, predicted results with empty array of results and finally both of them combined for further analyses of data.
## Heading 1.1
In below cell: Iterating over the json files and creating two dataframes. df_predicted for the json files returned as an array of results and df_empty returned as an empty array of results by the algorithm. 

```python
def get_biggest_rect(d):
    poly_array = []
    for c in d['results']:
        points = c['coordinates']
        po_array = []

        for xydict in points:
            po = Point(xydict['x'], xydict['y'])
            po_array.append(po)
        coords = [(pi.x,pi.y) for pi in po_array]
        poly = Polygon(coords)
        poly_array.append(poly.area)
    #                 print(poly_array)
    index = np.argmax(poly_array)
        #  for p ,take each x and y store in array of points                    
        # create polygon and find area, put in a array of area
    #               #max of area array and find position
    plate = d["results"][index]["plate"]
    conf = d["results"][index]["confidence"]    
        
            
    return (plate,conf)        
```

```python
os.chdir('/home/waqas/Desktop/data_upload/')

filelist = []

for files in glob.glob("*.json"):
    filelist.append(files)

df_predicted = pd.DataFrame()
df_empty = pd.DataFrame()

for file in filelist:
    with open(file, 'r') as json_data:
        try:
            data = json.load(json_data)
             
            if (len(data["results"]) == 1):
                plate = data["results"][0]["plate"]
                conf = data["results"][0]["confidence"]
                df_predicted = df_predicted.append({'num_plate': plate,'conf': conf, 'filename': file.strip('.json')}, ignore_index=True)
                
            elif (len(data["results"]) > 1):
                plate, conf = get_biggest_rect(data)
                df_predicted = df_predicted.append({'num_plate': plate,'conf': conf, 'file_name': file.strip('.jpg.json')}, ignore_index=True)

            elif (len(data["results"]) == 0):
                plate1 = data["results"]
                conf1 = data["results"]
                df_empty = df_empty.append({'num_plate': plate1,'conf': conf1, 'file_name': file.strip('.json')}, ignore_index=True)
        except Exception as e:
            print(e)
```
