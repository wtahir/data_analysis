```python
import pandas as pd
import csv
import glob
import os
```

# Heading 
Looping through ground truthed txt files and constructing a dataframe. It has four columns with 'extra' being a column containing the second number plate for the same image.

```python
os.chdir('/home/waqas/sandbox/bas/bas1/input/gt/')

filelist = []

for f in glob.glob('*.txt'):
    filelist.append(f)

df_gt = pd.DataFrame()
for file in filelist:
    frame = pd.read_csv(file,header=None,index_col=None)
    frame['filename'] = file[0:].strip('.jpg.txt|/')
    df_gt = df_gt.append(frame)

df_gt.columns = ('type', 'num_plate','filename','extra')
columns1 = ['filename', 'num_plate', 'type', 'extra']
df_gt = df_gt.reindex(columns=columns1)
df_gt = df_gt.drop(['extra'], axis=1)
df_gt['num_plate'] = df_gt['num_plate'].str.replace(" ","")
df_gt
```

```python
df_gt = df_gt[df_gt.type != 'n']
df_gt = df_gt[df_gt.type != 'c']
df_gt.head()
```

```python
df_gt = df_gt.drop(['num_plate', 'type'], axis=1)
df_gt.head()
```

```python
# df_gt.to_csv('without_c_n.lst', header=None, index=None)
os.chdir('/home/waqas/sandbox/bas/bas1/input/')

df_gt_with_type_txt = pd.read_csv('gt_with_type.txt', sep='\t', header=None)
df_filtered = pd.read_csv('sel.lst',sep='.', header=None)

df_filtered.columns = ['filename', 'gar']
df_filtered = df_filtered.drop(['gar'], axis=1)
df_filtered

df_gt_with_type_txt.columns = ['filename', 'gt_num_plate', 'type']
df_final_with_type = df_gt_with_type_txt.merge(df_filtered, left_on='filename', right_on='filename', how='inner')

df_final_with_type.to_csv('filtered_gt_with_type.txt', sep='\t', header=None)
```

# Heading
Dropping 'extra' column as it is decided not to consider 2 number plates in single image. Later taking value counts for each type (e.g, perpective, upright, cut, rotated and null). The dataframe reutrns the total number of ground truthed files with respect to types.

Reading predicted results file and naming columns based upon filename, confidence score and number plate
