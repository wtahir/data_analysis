```python
import pandas as pd
import os
import shutil
import glob
```

```python
df_old = pd.read_csv('service.md5', sep='\ \ ', names=['md5', 'file_service'])
df_old.head()
```

```python
df_new = pd.read_csv('service1.md5', sep='\ \ ', names=['md5', 'file_service1'])
df_new.head()
```

```python
df = df_old.merge(df_new, how='inner', left_on='md5', right_on='md5')
name = '/home/waqas/avrios/'
df['file_service'] = df['file_service'].str.strip('./')
df['file_service'] = name + df['file_service']

df['file_service1'] = df['file_service1'].str.strip('./')
df['file_service1'] = name + df['file_service1']

df['file_service1'].to_csv('service1_ids.txt', sep=('\t'), index=False)
df['file_service'].to_csv('service_ids.txt', sep=('\t'), index=False)
```
